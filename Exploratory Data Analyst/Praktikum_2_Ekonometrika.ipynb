#**Model Linier**

##**Model awal**

```r
#model Regresi
modellinier <- lm(Y ~ X2 + X3 + X4 + X5,  data = data)
summary(modellinier)
```

###**Uji Asumsi Model Linier**

```r
install.packages("AER")
install.packages("corrplot")
install.packages("tseries")
```

```r
library("AER")
library("corrplot")
library("olsrr")
library("tseries")
library(lmtest)
library(stats)
```

####**Uji Multikolinearitas**

```r
#Uji Multikolinearitas 1
##Matriks Korelasi
corrplot(cor(data[,1:5]), method = "number")
```

```r
##ViF
ols_vif_tol(modellinier)
```

```r
#Uji Multikolinearitas 2
##nilai eigen
ols_eigen_cindex(modellinier)
```

####**Uji Heterokesdatisitas**

```r
#Uji Heteroskesdatisitas
## White test atau Uji Breusch-Pagan
bptest(modellinier, ~ X2 + X3 + X4 +X5, data =
data)
## White HC standard errors
coeftest(modellinier, vcov = vcovHC(modellinier, type = "HC1"))
##plot
par(mfrow = c(2, 2))
plot(modellinier)
```

####**Uji Normalitas**

```r
#Uji Normalitas
## Jarque-Bera test
hist(residuals(modellinier), breaks = 16, ylim = c(0, 10), col =
"lightgray")
jarque.bera.test(residuals(modellinier))
```

####**Uji Autokorelasi**

```r
#Uji Autokorelasi
##Durbin-Watson Test
dwtest(modellinier)
## Test for Autocorrelation with the ACF Plot
acf(modellinier$residuals, type = "correlation", main = "Plot Autokorelasi model linier")
##Breusch-Godfrey Test
bgtest(modellinier, order = 3)
```

####**Deteksi kesalahan model regresi**

```r
#Uji Ramsey model linier awal
resettest(modellinier)
```

```r
#Metode Forward dan Backward
ols_step_forward_p(modellinier)
ols_step_backward_p(modellinier)
```

#**Perbandingan Semua Model**

```r
# Simpan summary dari setiap model
summary_model1 <- summary(modellinier)
summary_model2 <- summary(modelliniersw)
summary_model3 <- summary(modelln)
summary_model4 <- summary(modellnsw)

#Pendefinisian
aic_values <- c(AIC(modellinier), AIC(modelliniersw), AIC(modelln), AIC(modellnsw))
bic_values <- c(BIC(modellinier), BIC(modelliniersw), BIC(modelln), BIC(modellnsw))
msep_values <- c(mean((modellinier$residuals)^2), mean((modelliniersw$residuals)^2), mean((modelln$residuals)^2), mean((modellnsw$residuals)^2))
apc_values <- c(AIC(modellinier, k = length(modellinier$residuals)), AIC(modelliniersw, k = length(modelliniersw$residuals)), AIC(modelln, k = length(modelln$residuals)), AIC(modellnsw, k = length(modellnsw$residuals)))
fpe_values <- c(length(modellinier$residuals) * mean((modellinier$residuals)^2),length(modelliniersw$residuals) * mean((modelliniersw$residuals)^2), length(modelln$residuals) * mean((modelln$residuals)^2), length(modellnsw$residuals) * mean((modellnsw$residuals)^2))

# Buat DataFrame untuk perbandingan
comparison_df <- data.frame(
  Model = c("Model 1", "Model 2", "Model 3", "Model 4"),
  Multiple_Rsquared = c(summary_model1$r.squared, summary_model2$r.squared, summary_model3$r.squared, summary_model4$r.squared),
  Adjusted_Rsquared = c(summary_model1$adj.r.squared, summary_model2$adj.r.squared, summary_model3$adj.r.squared, summary_model4$adj.r.squared),
  F_Statistic = c(summary_model1$fstatistic[1], summary_model2$fstatistic[1], summary_model3$fstatistic[1], summary_model4$fstatistic[1]),
  AIC = aic_values,
  BIC = bic_values,
  MSE = msep_values,
  APC = apc_values,
  FPE = fpe_values
)

# Tampilkan DataFrame perbandingan
print(comparison_df)



```
