# Metode Jacobi dengan tebakan awal $x^0 = [0,0,0]$

```python
import numpy as np
from tabulate import tabulate

augmented_matrix = np.array(eval(input('Masukkan augmented matrix yang akan diselesaikan: ')))        #masukkan augmented matriks dari SPL
initial_solution = np.array(eval(input('Masukkan tebakan solusi awal: ')), dtype=float)               #masukkan tebakan awal solusi
tolerance = eval(input('Masukkan toleransi: '))                                                       #masukkan toleransi
A = np.array(augmented_matrix)                                                                        #mengubah list menjadi array matriks

def Jacob(matrix,initial,tol):                                                                        #pendefinisian Metode Jacob
  k = np.shape(matrix)[0]                                                                             #menentukan ukuran baris dan kolom matriks
  x0 = initial
  x = x0.copy()                                                                                       #agar x tidak ikut berubah
  for i in range(k):
    b = matrix[i,k]                                                                                   #konstanta spl
    for j in range(k):
      if j != i:#pendefinisian Metode Jacob
        b -= matrix[i,j]*x0[j]                                                                        #rumus utama Metode Jacob
    x[i]= b/matrix[i,i]
  banyak_iterasi = 1                                                                                  #pendefinisian iterasi awal yaitu dimulai dari 1
  table = []
  while np.linalg.norm(x-x0,np.inf)>=tol and banyak_iterasi < 4000 :                                  #penyetop error mutlak
    x0 = x #definisikan x0 sebagai x(n-1)
    x = x0.copy()                                                                                     #definisikan x sebagai x(n), tetapi awalnya disamakan dengan x0
    table.append([banyak_iterasi,x[0],x[1],x[2]])                                                     #masukkan elemen-elemen pada langkah iterasi ke sebuah list
    banyak_iterasi +=1
    for i in range(k):
      b = matrix[i,k]
      for j in range(k):
        if j!=i:
          b-=matrix[i,j]*x0[j]
      x[i]=b/matrix[i,i]

  table.append([banyak_iterasi,x[0],x[1],x[2]])                                                       #memasukkan hasil aproksimasi pada langkah akhir iterasi dalam tabel
  judul_label = ['Iterasi ke-','x1','x2','x3']
  print(tabulate(table, headers = judul_label , tablefmt = 'pretty', numalign = "center", stralign = "center", floatfmt = ".8f")) #mencetak tabel dari iterasi penyelesaian SPL
  print('Aproksimasi solusi SPL dengan Metode Jacobi adalah: ')
  return x

print(A)
print(Jacob(augmented_matrix,initial_solution,tolerance))                                                #mencetak aproksimasi penyelesaian SPL
```

# Metode Gauss Seidel dengan tebakan awal $x^0 = [0,0,0]$

```python
import numpy as np                                                                                    #import package
from tabulate import tabulate

augmented_matrix = np.array(eval(input('Masukkan augmented matrix yang akan diselesaikan: ')))        #masukkan augmented matriks dari SPL
initial_solution = np.array(eval(input('Masukkan tebakan solusi awal: ')), dtype=float)               #masukkan tebakan awal dari solusi spl
tolerance = eval(input('Masukkan toleransi: '))                                                       #Mengubah list menjadi array matriks
A = np.array(augmented_matrix)

def GaussSeidel(matrix,initial,tol):                                                                  #pendefinisian Metode Gauss Seidel
  k = np.shape(matrix)[0]                                                                             # menentukan ukuran baris dan kolom dari matriks (k
  x0 = initial
  x = x0.copy()                                                                                       #agar x tidak berubah
  for i in range(k):
    b = matrix[i,k]                                                                                   #konstanta dari SPL
    for j in range(k):
      if j != i:
        if j<i:
          b -= matrix[i,j]*x[j]
        else:
          b -= matrix[i,j]*x0[j]                                                                      #rumus Metode Gauss Seidel
    x[i]= b/matrix[i,i]
  banyak_iterasi = 1                                                                                  #pendefinisian iterasi dimulai dari 1
  tabel = []
  while np.linalg.norm(x-x0,np.inf)>=tol and banyak_iterasi < 4000 :                                  # penyetop error mutlak
    x0 = x                                                                                            #definisikan x0 sebagai x(n-1), dengan n banyak iterasi
    x = x0.copy()
    isi_label = [banyak_iterasi,x[0],x[1],x[2]]                                                       #pendefinisian x sebagai x(n), dengan n banyak iterasi
    tabel.append(isi_label)                                                                           #memasukkan elemen-elemen pada langkah iterasi ke sebuah list
    banyak_iterasi +=1
    for i in range(k):
      b = matrix[i,k]
      for j in range(k):
        if j!=i:
          if j<i:
            b -= matrix[i,j]*x[j]
          else:
           b -= matrix[i,j]*x0[j]                                                                     #rumus Metode Gauss Seidel
      x[i]= b/matrix[i,i]

  isi_label = [banyak_iterasi,x[0],x[1],x[2]]
  tabel.append(isi_label)                                                                             #memasukkan hasil aproksimasi pada langkah akhir iterasi dalam tabel
  judul_label = ['Iterasi ke-', 'x1', 'x2', 'x3']
  print(tabulate(tabel, headers = judul_label, tablefmt = 'pretty', numalign = "center", stralign = "center", floatfmt = ".8f")) #menampilkan tabel iterasi aproksimasi soluasi spl
  print('Aproksimasi solusi SPL dengan Metode Gauss Seidel adalah: ')                                 #mencetak aproksimasi solusi spl gauss seidel
  return x

print(A)                                                                                              #mencetak array matriks
print(GaussSeidel(augmented_matrix,initial_solution,tolerance))
```
