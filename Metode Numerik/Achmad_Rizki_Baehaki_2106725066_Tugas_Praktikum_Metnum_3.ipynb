Sebuah fungsi  f(x)  didefinisikan sebagai  x5+9x4−ax3+bx2−cx+5=0  dimana a, b, dan c adalah tiga digit terakhir berturut-turut dari NPM kalian. (Misal : untuk NPM 2006522123 maka fungsi yang digunakan adalah  x5+9x4−1x3+2x2−3x+5=0 ). Buatlah program yang dapat memberikan output berupa :

• Jika user memilih diturunkan, tampilkan

o TPMP

o TPEP

o FPMP

o FPEP

• Jika user memilih diintegralkan, tampilkan

o Metode Trapezoid

o Metode Simpson

• Jika user memilih diintegralkan komposit, tampilkan

o Metode Trapezoid Komposit

o Metode Simpson Komposit

Fungsi  : $x^5 + 9x^4 + 6x^2 -6x+ 5 = 0$

```python
from tabulate import tabulate
```

```python
input_formula = input('Masukkan formula fungsi yang akan dicari nilai turunannya : ')   #memasukkan formula fungsi
def f(x):
    return eval(input_formula)

#Rumus untuk tiap metode dari turunan numerik
def TPEP(x,h):
  return (1/(2*h))*(-3*f(x)+4*f(x+h)-f(x+2*h))
def TPMP(x,h):
  return (1/(2*h))*(f(x+h)-f(x-h))
def FPEP(x,h):
  return (1/(12*h))*(-25*f(x)+48*f(x+h)-36*f(x+2*h)+16*f(x+3*h)-3*f(x+4*h))
def FPMP(x,h):
  return (1/(12*h))*(f(x-2*h)-8*f(x-h)+8*f(x+h)-f(x+2*h))

 #Rumus untuk tiap metode dari integral numerik
def Trapezoid(a,b):                                       #Formula untuk metode Trapezoid
  h = b-a
  return (h/2)*(f(a)+f(b))

def Simpson(a,b):                                         #Formula untuk metode Simpson
  h = (b-a)/2
  return (h/3)*(f(a)+4*f(a+h)+f(b))

#Rumus untuk tiap metode dari integral numerik komposit
def CompTrapezoid(a,b,n):                                 #Formula untuk metode Trapezoid Komposit
  h = (b-a)/n
  sumtrap = 0
  for j in range(1,n):
    sumtrap = sumtrap + f(a+j*h)
    return (h/2)*(f(a)+2*sumtrap+f(b))

def CompSimpson(a,b,n):                                   #Formula untuk metode Simpson Komposit
  if n%2==1:
    return "jumlah partisi harus genap"
  elif n%2==0:
    h = (b-a)/n
    sum1 = 0
    sum2 = 0
    for j in range (1, int(n/2)):
      sum1+= f(a+2*j*h)
      sum2+= f(a+(2*j-1)*h)
      sum2 += f(b-h)
      return (h/3)*(f(a)+2*sum1+4*sum2+f(b))

print('Selamat datang diprogram Turunan, Integral, dan Integral Komposit')

#Tabel Penjelasan Program
table = [[1, "Turunan Numerik", "Mengaproksimasi nilai turunan fungsi pada suatu titik"],
         [2, "Integral Numerik", "Mengaproksimasi nilai integral fungsi pada batas interval (a,b)"],
         [3, "Integral Numerik Komposit", "Integral Numerik dengan Partisi"]]
print(tabulate(table,headers = ["Kode", "Program", "Fungsi"],tablefmt ="orgtbl"))

#Memberi penjelasan ulang dalam menjalankan program
Running = "Yes"
while Running == "Yes" or Running == "yes":
  input_metode = int(input("Pilih metode yang anda inginkan: \n(1) Turunan Numerik (2)Integral Numerik (3)Integral Numerik Komposit"))
  print(f"Metode yang anda pilih adalah metode: {input_metode}")
  while input_metode < 1 or input_metode > 3:
      print("Nomor metode yang anda tentukan tidak tersedia".format(input_metode))
      input_metode = int(input("Pilih metode yang anda inginkan: \n(1) Turunan Numerik (2)Integral Numerik (3)Integral Numerik Komposit"))
      print(f"Metode yang anda pilih adalah metode: {input_metode}")

  else:
    if input_metode == 1:
      x0 = eval(input('Masukkan titik dimana nilai turunan fungsinya akan diaproksimasi: ')) #titik dimana nilai turunan fungsi akan diaproksimasi
      h = eval(input('Masukkan besar stepsize: '))                                           #Besar Stepsize
      tabel = input("Output berupa tabel? (Yes/No): ")                                       #pilihan output berupa tabel
      if tabel == "Yes" or tabel == "yes":
        print("Turunan fungsi di x0 = {0} adalah: ".format(x0))
        a = TPEP(x0,h)
        b = TPMP (x0,h)
        c = FPEP(x0,h)
        d = FPMP(x0,h)
        hasiltabel= [["Metode Turunan Numerik","Aproksimasi"],["TPEP",a],["TPMP",b],["FPEP",c],["FPMP",d]]
        print(tabulate(hasiltabel, headers = "firstrow", tablefmt = "pretty"))
      elif tabel == "No" or tabel == "no":
        print('Turunan dari f di x0 = {0} adalah : '.format(x0))
        print("f'({0}) = {1:.7f} (Three Point Endpoint)".format(x0,TPEP(x0,h)))
        print("f'({0}) = {1:.7f} (Three Point Midpoint)".format(x0,TPMP(x0,h)))
        print("f'({0}) = {1:.7f} (Five Point Endpoint)".format(x0,FPEP(x0,h)))
        print("f'({0}) = {1:.7f} (Five Point Midpoint)".format(x0,FPMP(x0,h)))
      print('================================================')
      Running = str(input("Apakah anda mau menghitung ulang? (Yes/No): ")) #Pilihan untuk mecoba ulang program
      print('================================================')

    elif input_metode == 2:
      low_bound = eval(input('Masukkan batas bawah integral : '))     #batas bawah integral
      up_bound = eval(input('Masukkan batas atas integral : '))       #batas atas integral
      tabel = input("Output berupa tabel? (Yes/No): ")                #pilihan output berupa tabel
      if tabel == "Yes" or tabel == "yes":
        print("Integral fungsi di x0 = {0} adalah: ".format(input_formula))
        a = Trapezoid(low_bound,up_bound)
        b = Simpson(low_bound,up_bound)
        hasiltabel = [["Metode Integral Numerik","Aproksimasi"],["Trapezoid",a],["Simpson",b]]
        print(tabulate(hasiltabel, headers = "firstrow", tablefmt = "pretty"))
      elif tabel == "No" or tabel == "no":
        print("Hasil integral dari fungsi f(x) = {0} pada interval [{1},{2}] adalah : ".format(input_formula, low_bound, up_bound))
        print("Trapezoid = {:.7f}".format(Trapezoid(low_bound,up_bound)))
        print("Simpson = {:.7f}".format(Simpson(low_bound,up_bound)))
      print('================================================')
      Running = str(input("Apakah anda mau menghitung ulang? (Yes/No): ")) #Pilihan untuk mecoba ulang program
      print('================================================')

    elif input_metode == 3:
      low_bound = eval(input('Masukkan batas bawah integral : '))     #batas bawah integral
      up_bound = eval(input('Masukkan batas atas integral : '))       #batas atas integral
      partisi = eval(input('Masukkan jumlah partisi (n): '))          #partisi yang akan dimasukkan
      tabel = input("Output berupa tabel? (Yes/No): ")                #pilihan output berupa tabel atau tidak
      if tabel == "Yes" or tabel == "yes":
        print("Turunan fungsi di x0 = {0} adalah: ".format(input_formula))
        a = CompTrapezoid(low_bound,up_bound,partisi)
        b = CompSimpson(low_bound,up_bound,partisi)
        hasiltabel = [["Metode Integral Numerik","Aproksimasi"],["Trapezoid",a],["Simpson",b]]
        print("Hasil integral numerik komposit dari fungsi f(x) = {0} pada interval [{1},{2}] dengan {3} partisi adalah: ".format(input_formula, low_bound, up_bound, partisi))
        print(tabulate(hasiltabel, headers = "firstrow", tablefmt = "pretty"))
      elif tabel == "No" or tabel == "no" :
        print("Hasil integral dari fugsi f(x) = {0} pada interval [{1},{2}] adalah: ".format(input_formula, low_bound, up_bound))
        print("Composite Trapezoid dengan {0} partisi = {1:.5f}".format(partisi, CompTrapezoid(low_bound, up_bound, partisi)))
        print("Composite Simpson dengan {0} partisi = {1:.5f}".format(partisi, CompSimpson(low_bound, up_bound, partisi)))
      print('================================================')
      Running = str(input("Apakah anda mau menghitung ulang? (Yes/No): ")) #Pilihan untuk mecoba ulang program
      print('================================================')

if Running == "No"  or Running == "no" :                              #Program berhenti
  print("-Terimakasih-")

```
