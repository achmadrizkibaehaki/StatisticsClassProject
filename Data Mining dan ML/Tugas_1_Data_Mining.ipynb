```python
# URL ke berkas Excel di GitHub
url = 'https://raw.githubusercontent.com/achmadrizkibaehaki/DataMining-BI/main/Tugas1/Data-Property-Indonesia-Ganjil-2023.xlsx'

# Membaca semua lembar kerja dari berkas Excel dan menyimpannya dalam bentuk dictionary
df = pd.read_excel(url, sheet_name=None)

# Mengidentifikasi nama kolom untuk setiap lembar kerja
for sheet_name, sheet_data in df.items():
    nama_kolom = sheet_data.columns.tolist()
    dimensi = sheet_data.shape
    print("Nama-nama kolom dalam lembar kerja '{}':".format(sheet_name))
    print(nama_kolom)
    print(dimensi)
```

```python
# URL ke berkas Excel di GitHub
url = 'https://raw.githubusercontent.com/achmadrizkibaehaki/DataMining-BI/main/Tugas1/Data-Property-Indonesia-Ganjil-2023.xlsx'

# Membaca semua lembar kerja dari berkas Excel
df = pd.read_excel(url, sheet_name=None)

# Nama kolom baru yang diinginkan
nama_kolom_baru = ['created_at', 'LT', 'LB', 'KT', 'KM', 'garasi', 'carport', 'lokasi', 'sertifikat', 'listrik', 'hadap', 'harga', 'URL', 'deskripsi']

# Mengubah nama kolom di setiap lembar kerja
for sheet_name, sheet_data in df.items():
    # Menghapus spasi tambahan di awal atau akhir nama kolom dan mengonversi ke huruf kecil
    sheet_data.columns = sheet_data.columns.str.strip().str.lower()
    # Menggunakan metode rename untuk mengubah nama kolom sesuai dengan nama_kolom_baru
    sheet_data.rename(columns=dict(zip(sheet_data.columns, nama_kolom_baru)), inplace=True)

    print(f"Nama kolom di lembar kerja '{sheet_name}' telah diubah menjadi: {nama_kolom_baru}")
```

```python
import pandas as pd
import numpy as np

# Fungsi untuk membersihkan nilai pada kolom power
def clean_power_values(value):
    if isinstance(value, str):
        # Menghapus 'Watt', 'watt', 'Va/R-1', 'Token', '-' dan mengganti 'KWh' dengan '000'
        cleaned_value = value.lower().replace('watt', '').replace('va/r-1', '').replace('token', '').replace('-', '').replace('kwh', '000')
        # Menghapus karakter selain angka
        cleaned_value = ''.join(filter(str.isdigit, cleaned_value))
        if cleaned_value == '':  # Jika string hanya terdiri dari karakter selain angka
            return np.nan
        else:
            return int(cleaned_value)
    return value

# Loop melalui semua sheet dan membersihkan kolom 'power'
for sheet_name in df.keys():
    df[sheet_name]['listrik'] = df[sheet_name]['listrik'].apply(clean_power_values)

# Outputkan DataFrame setelah pengolahan
for sheet_name, sheet_df in df.items():
    print(f"Sheet: {sheet_name}")
    print(sheet_df['listrik'])
```

```python
#Detect outliers
import seaborn as sns
import matplotlib.pyplot as plt
columns = merged_df.columns.values
for i in columns:
    if merged_df[i].dtype=='float64'or merged_df[i].dtype=='int64':
       sns.boxplot(merged_df[i])
       plt.title(i)
       plt.show()
```

```python
import pandas as pd

# Assuming df is your DataFrame
# Select only numeric columns
numeric_columns = merged_df.select_dtypes(include=['number']).columns

# Dictionary to store outlier dataframes for each column
outlier_dataframes = {}

# Dictionary to store non-outlier dataframes for each column
non_outlier_dataframes = {}

# Iterate through numeric columns to separate outliers and non-outliers
for column_name in numeric_columns:
    # Calculate the IQR (Interquartile Range)
    Q1 = merged_df[column_name].quantile(0.25)
    Q3 = merged_df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds for outlier detection (3 times IQR from median)
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR

    # Separate outliers and non-outliers
    outliers = merged_df[(merged_df[column_name] < lower_bound) | (merged_df[column_name] > upper_bound)]
    non_outliers = merged_df[(merged_df[column_name] >= lower_bound) & (merged_df[column_name] <= upper_bound)]

    # Store outlier and non-outlier dataframes in the dictionaries
    outlier_dataframes[column_name] = outliers
    non_outlier_dataframes[column_name] = non_outliers

# Print the first few rows of each outlier dataframe
for column_name, outlier_df in outlier_dataframes.items():
    print(f"Outliers in column '{column_name}':")
    print(outlier_df.head())

# Print the first few rows of each non-outlier dataframe
for column_name, merged_df in non_outlier_dataframes.items():
    print(f"Non-outliers in column '{column_name}':")
    print(merged_df.head())
```
